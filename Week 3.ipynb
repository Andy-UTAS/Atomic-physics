## Visualisation of states

### State definitions 

```python
def isw_state(n, x, L):
    """
    isw_state specifies the energy eigenstates of the infinite square well

    :param n: Energy level
    :param x: The position [m]
    :param L: Width of the well [m]
    :return: Value of the wavefunction
    """
    
    return np.sqrt(2/L) * np.sin(n * pi * x / L)
    
def isw_dist_prob(n1, n2, x1, x2, L):
    """
    isw_state returns the probability density for distinguishable 2-particle state for the infinite square well

    :param n: Energy level
    :param x: The position [m]
    :param L: Width of the well [m]
    :return: The form of the differential equation
    """
    
    return (np.abs(isw_state(n1, x1, L) * isw_state(n2, x2, L))) ** 2

def isw_bose_prob(n1, n2, x1, x2, L):
    """
    isw_state returns the probability density for a given 2-particle state for the infinite square well

    :param n: Energy level
    :param x: The position [m]
    :param L: Width of the well [m]
    :return: The form of the differential equation
    """
    
    return (np.abs(1/np.sqrt(2) * (isw_state(n1, x1, L) * isw_state(n2, x2, L) + isw_state(n1, x2, L) * isw_state(n2, x1, L)))) ** 2

def isw_fermi_prob(n1, n2, x1, x2, L):
    """
    isw_state returns the probability density for a given 2-particle state for the infinite square well

    :param n: Energy level
    :param x: The position [m]
    :param L: Width of the well [m]
    :return: The form of the differential equation
    """
    
    return (np.abs(1/np.sqrt(2) * (isw_state(n1, x1, L) * isw_state(n2, x2, L) - isw_state(n1, x2, L) * isw_state(n2, x1, L)))) ** 2
```

## Exchange interaction

Compute the probability density for the exchange interaction for the first excited state of the infinite square well

```python
def ips_d(u, L):
    """
    ips_d interparticle separation for distinguiable partilces
    
    :param u: interparticle separation, scaled by pi/L
    :param L: Width of the well [m]
    :return: Probability density
    """
        
    return (1/L**2) * (pi - abs(u) + 2/3 * np.sin(2*abs(u)) - 1/12 * np.sin(4*abs(u)))

def ips_i(u, L):
    """
    ips_i interparticle separation for indistinguishable partilces
    
    :param u: interparticle separation, scaled by pi/L
    :param L: Width of the well [m]
    :return: Probability density
    """
    
    t1 = np.cos(2*u)
    t2 = 2 * np.cos(u)
    t3 = 9 * np.sin(abs(u))
    t4 = np.sin(3*abs(u))
    
    s_1 = (1/(3*L ** 2)) * ((pi - abs(u)) * (12 + 6*(t1 - t2)) - t3 + 9 * np.sin(2*abs(u)) - t4) * (np.cos(u/2)) ** 2
    s_2 = (1/(3*L ** 2)) * ((pi - abs(u)) * (12 + 6*(t1 + t2)) + t3 + 9 * np.sin(2*abs(u)) + t4) * (np.sin(u/2)) ** 2
                            
    return s_1, s_2
```

```python
L = 1

# Create the x values over which the potential will be plotted
x_vals = np.arange(-L, L, 0.01) # x-values of the peak

# Plot the spectrum
fig, ax = plt.subplots(figsize=(10,8)) # Initialise the plot

ax.plot(x_vals, ips_d(x_vals * pi/L, L), '--', label='Distinguisable') # plot the potential

identical = ips_i(x_vals * pi/L, L)

ax.plot(x_vals, identical[0], label='Symmetric') # plot the potential
ax.plot(x_vals, identical[1], label='Antisymmetric') # plot the potential


plt.xlabel('$x_1 - x_2$ [L]') # Label the x axis
plt.ylabel('$\mathscr{P}(x_1 - x_2)$') # Label the y axis
plt.setp(ax.get_yticklabels(), visible=False)
ax.tick_params(axis='y', length=0)
plt.locator_params(axis='x', nbins=6)
plt.title('Probability density of interparticle separation')
plt.legend()

if savefigflag: # Change to True to save the figures
    plt.savefig('Images/03-01_07-interparticle.svg', bbox_inches= 'tight')

plt.show()
```

## Perturbation theory

```python
def prob(t, w_fi, w):
    """
    isw_state specifies the energy eigenstates of the infinite square well

    :param n: Energy level
    :param x: The position [m]
    :param L: Width of the well [m]
    :return: Value of the wavefunction
    """
    
    const = 1
    
    return const * (np.sin(t * (w_fi - w)/2) ** 2/((w_fi - w)/2)**2)

w_fi = 4
w_max = 10

# Create the x values over which the potential will be plotted
w_vals = np.arange(.001, w_max, 0.01)

# Plot the spectrum
fig, ax = plt.subplots(figsize=(10,8)) # Initialise the plot

ax.plot(w_vals, prob(4, w_fi, w_vals)) # plot the potential


plt.xlabel('$\omega$') # Label the x axis
plt.ylabel('$\mathscr{P}(\omega)$') # Label the y axis
plt.setp(ax.get_yticklabels(), visible=False)
ax.tick_params(axis='y', length=0)
plt.setp(ax.get_xticklabels(), visible=False)
ax.tick_params(axis='x', length=0)
plt.locator_params(axis='x', nbins=6)
plt.title('Excitation probability')

if savefigflag: # Change to True to save the figures
    plt.savefig('Images/03-03_01-interparticle.svg', bbox_inches= 'tight')

plt.show()
```

### Blackbody

```python
def planck(wavelength, temperature):
    """
    Calculate the blackbody radiation spectrum using Planck's law.
    
    Parameters:
    wavelength (array): Wavelengths in meters.
    temperature (float): Temperature in Kelvin.
    
    Returns:
    array: Spectral radiance (W/sr/m^3)
    """
    a = 2.0 * h * c**2
    b = h * c / (wavelength * kb * temperature)
    intensity = a / ( (wavelength**5) * (np.exp(b) - 1.0) )
    return intensity

# Wavelength range (in meters)
wavelengths = np.linspace(1e-8, 3e-6, 1000)

# Temperature (in Kelvin)
temperatures = [3000, 4000, 5000, 6000]

# Plotting
plt.figure(figsize=(10, 6))
for T in temperatures:
    intensity = planck(wavelengths, T)
    plt.plot(wavelengths * 1e9, intensity, label=f'T = {T} K')

plt.title('Blackbody Radiation Spectrum')
plt.xlabel('Wavelength [nm]')
plt.ylabel('Spectral Radiance [W/sr/m$^3$]')
plt.legend()
plt.grid(True)

if savefigflag: # Change to True to save the figures
    plt.savefig('Images/03-03_02_blackbody.svg', bbox_inches= 'tight')

plt.show()
```

```python
def lorentzian(omega, a_21, omega_21):
    """
    Calculate the blackbody radiation spectrum using Planck's law.
    
    Parameters:
    omega (array): Wavelengths in meters.
    a_21 (float): Temperature in Kelvin.
    omega_21 (float):
    
    Returns:
    array: Spectral radiance (W/sr/m^3)
    """
    
    return ((a_21)/(2 * pi))/((omega - omega_21) ** 2 + (a_21/2) ** 2)

a_21 = 4e13
omega_21 = 2*pi*100e12

I = .1
B_12 = 1

# Create the x values over which the potential will be plotted
omeag_vals = np.linspace(1, 1e15, 1000)
rate_vals = 3 * (I/c) * B_12 * lorentzian(omeag_vals, a_21, omega_21)

# Plot the spectrum
fig, ax = plt.subplots(figsize=(10,8)) # Initialise the plot

ax.plot(omeag_vals, rate_vals) # plot the potential


plt.xlabel('Frequency') # Label the x axis
plt.ylabel('$R_{12}(\omega)$') # Label the y axis
plt.setp(ax.get_yticklabels(), visible=False)
ax.tick_params(axis='y', length=0)
ax.set_xticks([omega_21])
ax.set_xticklabels(["$\omega_{21}$"])
plt.title('Excitation probability')

if savefigflag: # Change to True to save the figures
    plt.savefig('Images/03-03_03-lorentzian.svg', bbox_inches= 'tight')


plt.show()
```
