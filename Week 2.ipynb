## The hyperfine effect

```python
def projection_angles(l):
    mj = np.linspace(-l,l,2*l+1)
    proj =[]
    for m in mj:
        angle = np.sign(m) * np.arccos(abs(m)/np.sqrt(l*(l+1)))
        proj.append((m, angle))
    return np.array(proj)

def plot_proj(l):
    r = np.sqrt(l*(l+1))
    pr = projection_angles(l)
    
    xy = r*np.sin(abs(pr[:,1])), np.sign(pr[:,1])*r*np.cos(pr[:,1])
    
    fig = plt.figure(figsize = (10, 10))
    ax = fig.add_subplot(polar = False)
    arr_width = .025
    ax.arrow(0,0, r, 0, length_includes_head = True, width = arr_width, head_width = 7.5 * arr_width, 
              head_length = 5 * arr_width)
    for v in np.transpose(xy):
        ax.arrow(0,0, v[0], v[1], length_includes_head = True, width = arr_width, head_width = 7.5 * arr_width, 
              head_length = 5 * arr_width)
    ax.set_aspect(1)
    
    ax.add_patch(plt.Circle((0, 0), r, fill=False))
    ax.set_xlim(0,r)
    
    plt.xlabel('$J_x~[\hbar]$') # Label the x axis
    plt.ylabel('$J_z~[\hbar]$') # Label the y axis
    plt.title('Angular momentum')
    
    if False: # Change to True to save the figures
        plt.savefig('w2_Angular-momentum.svg', bbox_inches= 'tight') # Name each figure by the lamp element
        
    plt.show()
    
plot_proj(1)
```
